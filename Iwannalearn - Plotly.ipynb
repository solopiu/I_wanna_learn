```python
import requests 
import pandas as pd 
```

```python
url = "https://opendata.ecdc.europa.eu/covid19/casedistribution/json/"

json_url = requests.get(url) 
json_file = json_url.json()
df = pd.DataFrame.from_dict(json_file['records'])
```

```python
df.head()
```

```python
df.info()
```

```python
ita = df.loc[df['countriesAndTerritories']=="Italy",:]
ita = ita[::-1]
ita.head()
```

```python
import plotly.graph_objs as go
from plotly.offline import iplot
# Creating trace1
trace1 = go.Scatter(
                    x = ita.dateRep,
                    y = ita.deaths,
                    mode = "lines+markers",
                    name = "deaths",
                    marker=dict(
                            color='Red',
                            size=5,
                            line=dict(
                                color='grey',
                                width=1
                            )),
                   # marker = dict(color = 'rgba(153, 0, 153, 0.3)'),
                    text= ita.cases)
# Creating trace2
trace2 = go.Scatter(
                    x = ita.dateRep,
                    y = ita["Cumulative_number_for_14_days_of_COVID-19_cases_per_100000"],
                    mode = "lines+markers",
                    name = "Cumulative_number_for_14_days_of_COVID-19_cases_per_100000",
                    marker = dict(color = 'rgba(80, 26, 80, 0.3)'),
                    text= ita.cases)
data = [trace1, trace2]
layout = dict(title = 'Cases and deaths for covid 19',
              xaxis= dict(title= 'Dates',ticklen= 5,zeroline= False)
             )
fig = dict(data = data, layout = layout)
iplot(fig)
```

```python
import plotly.graph_objs as go
# create trace1 

group = df.groupby('countriesAndTerritories', as_index=False).agg({'deaths':'sum', 'cases':'sum'})
trace1 = go.Bar(
                x = group.countriesAndTerritories,
                y = group.deaths,
                name = "deaths",
                yaxis="y1",
                marker = dict(color = 'rgba(255, 174, 255, 0.5)',
                             line=dict(color='rgb(0,0,0)',width=1.5))
)
# create trace2 
trace2 = go.Scatter(
                x = group.countriesAndTerritories,
                y = group.cases,
                mode = "lines+markers",
                name = "cases",
                yaxis="y2",
                marker = dict(color = 'rgba(255, 0, 128, 0.5)',
                              line=dict(color='rgb(0,0,0)',width=1.5))
                )
data = [trace1, trace2]
layout = go.Layout(barmode = "group", 
                   yaxis=dict(title='Deaths'),
                       yaxis2=dict(title='Cases',
                                   type="log",
                                   overlaying='y',
                                   side='right'))
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```

# Testing data

```python
url = "https://opendata.ecdc.europa.eu/covid19/testing/json/"
json_url = requests.get(url) 
json_file = json_url.json()
test = pd.DataFrame.from_dict(json_file)
```

```python
test.head()
```

```python
test.info()
```

```python

# import graph objects as "go"
import plotly.graph_objs as go

test_ita = test.loc[test['country']=="Italy",:]
trace1 = {
  'x': test_ita.year_week,
  'y': test_ita.testing_rate,
  'name': 'testing rate',
  'type': 'bar',
    'yaxis':'y1'
};
trace2 = {
  'x': test_ita.year_week,
  'y': test_ita.positivity_rate,
  'name': 'positivity rate',
  'type': 'scatter',
    'marker':{'color':'rgba(255, 50, 96, 1)', 
               'size':1000,
             },
    'yaxis':'y2'
};
data = [trace1, trace2];
layout = {
  'xaxis': {'title': 'year_week'},
  'barmode': 'group',
  'title': 'testing rate and positivity rate in Italy',
    'yaxis':{'title':'Testing rate'},
    'yaxis2':{'title':'Positivity rate', 'overlaying':'y', 'side':'right'}
};

fig = go.Figure(data = data, layout = layout)
iplot(fig)
```

```python
test['positivity_rate']=test['positivity_rate'].astype(float)
test2 = test.groupby('country', as_index=False).agg({'positivity_rate':'mean'})
# figure
fig = {
  "data": [
    {
      "values": test2.positivity_rate,
      "labels": test2.country,
      #"domain": {"x": [0, .5]},
      "name": "",
      "hoverinfo":"label+percent+name",
      "hole": .3,
      "type": "pie"
    },],
  "layout": {
        "title":"Average Positivity rate by country",
        "annotations": [
            { "font": { "size": 20},
              "showarrow": False,
              "text": "",
                "x": 0.20,
                "y": 1
            },
        ]
    }
}
iplot(fig)
```

```python
test['testing_rate']=test.testing_rate.astype(float)
test3 = test.groupby('year_week', as_index=False).agg({'positivity_rate':'mean',
                                                       'testing_rate':'mean',
                                                      'new_cases':'mean'})
data = [
    {
        'x': test3.year_week,
        'y': test3.testing_rate,
        'mode': 'markers',
        'marker': {
            'color':test3.new_cases,
            'size': test3.positivity_rate*3,
            'showscale': True
        },
        "text" :  "mean new cases"    
    }
]
iplot(data)
```

```python
trace0 = go.Box(
    y=test_ita.testing_rate,
    name = 'testing rate',
    marker = dict(
        color = 'rgb(12, 12, 140)',
    )
)
trace1 = go.Box(
    y=test_ita.new_cases,
    name = 'tests done',
    marker = dict(
        color = 'rgb(12, 128, 128)',
    )
)
data = [trace0, trace1]
iplot(data)
```
